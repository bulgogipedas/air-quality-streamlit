# Exploratory Data Analysis (EDA)

    Karena tabelnya memiliki column dan tipe data yang sama, kita bisa melakukan union terhadap semua tabel agar menambahkan baris sesuai masing-masing data.

```python
combine_df = pd.concat([dongsi_df, shunyi_df, wanliu_df], ignore_index=True)
combine_df
```

```python
# Mengubah semua nama kolom menjadi huruf kecil
combine_df.columns = map(str.lower, combine_df.columns)

# Menampilkan lima baris pertama data untuk memastikan perubahan
combine_df.head()
```

Merubah semua colomn menjadi lower agar semua kolom konsisten.

```python
combine_df.describe(include='all')
```

```python
# Define the ranges and AQI (Air Quality Index) upper bounds
category_ranges = [
    'Good', 'Moderate', 'Unhealthy for Sensitive Groups', 'Unhealthy', 'Very Unhealthy', 'Hazardous'
]
pm25_ranges = [12, 35.4, 150.4, 250.4, float('inf')]
pm10_ranges = [54, 154, 254, 354, 424, float('inf')]
so2_ranges = [35, 75, 185, 304, 604, float('inf')]
no2_ranges = [53, 100, 360, 649, 1249, float('inf')]
co_ranges = [4400, 9400, 12400, 15400, 30400, float('inf')]
o3_ranges = [54, 70, 85, 105, 200, float('inf')]

# Define a function to categorize a value based on the given ranges
def categorize(value, ranges):
    for max_val in ranges:
        if value <= max_val:
            return ranges.index(max_val)

# Create a new column 'Category' by applying the categorize function to each row
combine_df['Category'] = combine_df.apply(lambda row: category_ranges[max(
    categorize(row['pm2.5'], pm25_ranges),
    categorize(row['pm10'], pm10_ranges),
    categorize(row['so2'], so2_ranges),
    categorize(row['no2'], no2_ranges),
    categorize(row['co'], co_ranges),
    categorize(row['o3'], o3_ranges),
)], axis=1)
```

```python
# Menambahkan kolom baru 'rain_category' berdasarkan 'rain' dengan kategori yang sudah ditentukan
bins = [0, 0.5, 20, 50, 100, 150, float('inf')]
labels = ['Overcast', 'Light rain', 'Moderate rain', 'Heavy rain', 'Very heavy rain', 'Extreme rain']
combine_df['rain_category'] = pd.cut(combine_df['rain'], bins=bins, labels=labels, right=False)

combine_df
```

Ini menunjukkan bahwa intensitas hujan yang lebih tinggi dapat berkontribusi pada peningkatan polusi udara dalam beberapa kasus. Namun, perlu dicatat bahwa ada beberapa parameter, seperti O3 (ozon), yang tampaknya tidak menunjukkan tren yang sama; nilai rata-rata O3 malah menurun saat hujan lebih intensif. Ini menunjukkan bahwa hubungan antara hujan dan kualitas udara bisa kompleks dan tergantung pada banyak faktor.

# Data Viz dan Explonatory Analysis

### Pertanyaan 1: Bagaimana tren polusi udara (PM2.5, PM10, SO2, NO2, CO, O3) dari tahun ke tahun?

```python
# Menggunakan dt.year untuk mendapatkan tahun dari kolom date
years = combine_df['date'].dt.year.unique()
pollutants = ['pm2.5', 'pm10', 'so2', 'no2', 'co', 'o3']
yearly_averages = [combine_df.groupby(combine_df['date'].dt.year)[p].mean() for p in pollutants]

# Membuat visualisasi
plt.figure(figsize=(10, 6))
for i in range(len(pollutants)):
    plt.plot(years, yearly_averages[i], label=pollutants[i])

plt.title('Air Pollution Trends from Year to Year')
plt.xlabel('Years')
plt.ylabel('Average Air Pollution')
plt.legend()
plt.grid(True)
plt.show()
```

### Pertanyaan 2: Bagaimana pola harian dari tingkat polusi udara?

```python
plt.figure(figsize=(10, 6))

# Plotting menggunakan Seaborn
sns.lineplot(data=hourly_average_pm25, label='PM2.5')
sns.lineplot(data=hourly_average_pm10, label='PM10')
sns.lineplot(data=hourly_average_so2, label='SO2')
sns.lineplot(data=hourly_average_no2, label='NO2')
sns.lineplot(data=hourly_average_co, label='CO')
sns.lineplot(data=hourly_average_o3, label='O3')

plt.title('Daily Pattern of Air Pollution Levels')
plt.xlabel('Hour')
plt.ylabel('Average Air Pollution Level')
plt.legend()
plt.grid(True)
plt.show()
```

### Pertanyaan 3: Apakah terdapat korelasi antara suhu (TEMP) dan tingkat polusi udara?

```python
correlation_matrix = combine_df[['temp', 'pm2.5', 'pm10', 'so2', 'no2', 'co', 'o3']].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='rocket', fmt=".2f")
plt.title('Correlation between Temperature and Air Pollution Parameters')
plt.show()
```

    Ternyata temperatur tidak memiliki korelasi dengan polusi udara, temperatur hanya memiliki korelasi positif dengan O3.

### Pertanyaan 4: Bagaimana distribusi tingkat polusi udara di setiap stasiun pengamatan?

```python


# Dictionary mapping station names to colors
station_colors = {'Wanliu': 'blue', 'Shunyi': 'orange', 'Dongsi': 'green'}

plt.figure(figsize=(10, 6))
sns.histplot(data=combine_df, x='pm2.5', hue='station', kde=True, bins=20)  # Ubah jumlah bins di sini

# Creating custom legend
legend_labels = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=station) 
                 for station, color in station_colors.items()]
plt.legend(handles=legend_labels, title='Observation Station')

plt.title('Distribution of Air Pollution Levels (PM2.5) in Each Observation Station')
plt.xlabel('Air Pollution Level (PM2.5)')
plt.ylabel('Frequency')
plt.show()
```

### Pertanyaan 4: Bagaimana pengaruh hujan terhadap polusi udara?

```python
palette = sns.color_palette("rocket", len(rainy_average_df.columns))

fig, ax = plt.subplots(figsize=(10, 6))

num_bars = len(rainy_average_df.index)
bar_width = 0.15

index = np.arange(num_bars)

for i, (column, color) in enumerate(zip(rainy_average_df.columns, palette)):
    ax.bar(index + i * bar_width, rainy_average_df[column], bar_width, label=column, color=color)

# Set labels and title
ax.set_xlabel('Rain Condition')
ax.set_ylabel('Average Pollutant Level')
ax.set_title('Average Pollutant Levels by Rain Condition')

# Set x-axis labels
ax.set_xticks(index + (num_bars * bar_width) / 2)
ax.set_xticklabels(rainy_average_df.index)

# Add legend
ax.legend(title='Pollutant')

plt.tight_layout()
plt.show()
```

    Dari data diatas menunjukkan bahwa hampir seluruh pollutant mengalami fluktuasi dalam setiap kategori hujan dikarenakan tidak memiliki korelasi diantara pollutant dan hujan. Namun di pollutant `co` kita dapat melihat terdapat kenaikan di setiap tingginya curah hujan

### Pertanyaan 5: Bagaimana polusi udara PM2.5 dan PM10 di setiap stasiun?

```python
colors = ["#ce7e00", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
# Sort data by PM10 levels
pm10_sorted = combine_df.groupby('station')['pm10'].mean().sort_values(ascending=False)
pm25_sorted = combine_df.groupby('station')['pm2.5'].mean().sort_values(ascending=False)

# Create subplots for PM2.5 and PM10
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot PM2.5
sns.barplot(y=pm25_sorted.index, x=pm25_sorted.values, ax=axs[0], palette=colors)
axs[0].set_ylabel(None)
axs[0].set_xlabel(None)
axs[0].set_title('PM2.5 Levels at Each Observation Station')

# Plot PM10
sns.barplot(y=pm10_sorted.index, x=pm10_sorted.values, ax=axs[1], palette=colors)
axs[1].set_ylabel(None)
axs[1].set_xlabel(None)
axs[1].invert_xaxis()
axs[1].yaxis.set_label_position("right")
axs[1].yaxis.tick_right()
axs[1].set_title('PM10 Levels at Each Observation Station')


# Adjust layout
plt.tight_layout()
plt.show()
```
